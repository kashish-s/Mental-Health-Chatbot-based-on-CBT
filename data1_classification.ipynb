```python
vocab = []#creating an empty list to store vocab
tags = []#to store tags
sentences = []#to store sentences
#question mark adds nothing to context, etc
ignore = ['?','!',',','.',':','"']
#storing the list of dictionaries in a list

lis = my_data['intents']

#print(lis)

print(lis[0])

```

```python
for d in lis:

  for sent in d['patterns']:

    tokens = nltk.word_tokenize(sent)

    vocab.extend(tokens)


    sentences.append((tokens, d['tag']))

    if d['tag'] not in tags:

      tags.append(d['tag'])



```

```python
#Stemming and converting to lower case

vocab = [stemmer.stem(tokens.lower()) for tokens in vocab if tokens not in ignore]
vocab = sorted(list(set(vocab)))

 #iterating through all tokens in vocab, and if the token is not a word which has to be ignored, then it will be stemmed and added to vocab list
#set() - gets all unique values, creates a list of it and sorts as .set returns in reverse order

#sorting tags

tags  = sorted(list(set(tags)))
```

```python
#import random
import random
```

```python
train = []

op_empty = [0]*len(tags)

for s in sentences:

  bagofwords = []
  pattern_tokens = s[0]

  stem_pattern_tokens = []

  for token in pattern_tokens:

    stem_token = stemmer.stem(token)
    stem_pattern_tokens.append(stem_token)

  for word in vocab:

    if word in stem_pattern_tokens:
      bagofwords.append(1)
    else:
      bagofwords.append(0)

  op_row = list(op_empty)
  index = tags.index(s[1])

  op_row[index] = 1

  train.append([bagofwords, op_row])

random.shuffle(train)




  
#Xtrain is bag of words
#labels - tags of each sentence 
```

```python
train = np.array(train)
```

```python
#x - sentences - bag of words

X_train = list(train[:,0])
```

```python
 ## tags is testing data

 y_train = list(train[:,1])
```

```python
from keras.models import Sequential#Sequential model
from keras.layers import Dense, Dropout, Activation #types of layers 
#optimizer
#import tensorflow
import tensorflow as tf
```

```python
#checking length of training data
len(X_train[0])
```

```python
#Building the model
model = Sequential(
    
    [tf.keras.layers.Dense(128, input_shape = (len(X_train[0]),), activation = 'relu'), #input layer
    tf.keras.layers.Dropout((0.5)),   
    tf.keras.layers.Dense(64, activation = 'relu'),
    tf.keras.layers.Dropout((0.5)),
    tf.keras.layers.Dense(len(tags), activation = 'sigmoid')]
)
```

```python
#defining optimizer with lr = 0.01 , decay as 1e-6 and using nesterov SGD, #decay for reularization purposes
from keras.optimizers import SGD
sgd = SGD(lr = 0.01, decay = 1e-6, nesterov = True)
```

```python
#compiling model
model.compile( loss = 'categorical_crossentropy', optimizer = sgd, metrics = ['accuracy'])
```

```python
# we will have to import some libraries for that
from keras.callbacks import Callback
```

```python
# we will have to create a class called Callbacks and initialize loga as a dictionary

class myCallback(tf.keras.callbacks.Callback):

  def on_epoch_end(self, epoch, logs = {}):

    if(logs.get('accuracy')>0.989):

      print("\n Accuracy is too high stop training!")
      self.model.stop_training = True

```

```python
# Fit the model and store model history
callbacks = myCallback()
history = model.fit(np.array(X_train), np.array(y_train), epochs = 250, batch_size = 5, verbose = 1, callbacks=[callbacks])
```

```python
#function to clean sentences
def clean(statement):
  statement_tokens = nltk.word_tokenize(statement)
  statement_tokens = [stemmer.stem(token.lower()) for token in statement_tokens if token not in ignore]

  return statement_tokens
```

```python
#taking an example 

clean('Hello! what are you up to?')
```

```python
#function to create bag of words for each sentence  

def bag_of_words(statement, vocab):
  
  statement_tokens = clean(statement)
  bag = [0]*len(vocab)

  for token in statement_tokens:

    for i,word in enumerate(vocab):
      if word == token:
        bag[i] = 1


  return np.array(bag)
```
